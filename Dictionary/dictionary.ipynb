```python
#creating a dictionary from two lists

keys = ["id", "name", "age"]
values = [101, "Alice", 25]

result = dict(zip(keys, values))
print(result)  # {'id': 101, 'name': 'Alice', 'age': 25}
```

```python
student = {"name": "John", "age": 21, "marks": 88}

student["grade"] = "A"          # add
student["marks"] = 90           # update
student.pop("age")              # remove

print(student)  # {'name': 'John', 'marks': 90, 'grade': 'A'}
```
