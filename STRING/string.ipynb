```python
#reverse a string 

def reverse(s):
    result = ""

    for ch in s:
        result =  ch + result
    return result

print(reverse("Pushkar"))


```

```python
#find length of string without len function

def string_length(s):
    count = 0
    for _ in s:
        count +=1
    return count
print(f"length of string is {string_length("Pushkar")}")

```

```python
#check if two strings rotation of each other

def is_rotation(s1, s2):
    if len(s1) != len(s2):
        return False
    return s2 in (s1+s2)

print(is_rotation("abcde", "cdeab"))
```

```python
#count vowels and consonants in a string
def count(s):
    count_vowels = 0
    count_conso = 0

    for ch in s:
        if(ch == 'a' or ch == 'e' or ch == 'i' or ch =='o' or ch == 'u'):
            count_vowels += 1
        elif(ch != 'a' or ch != 'e' or ch != 'i' or ch !='o' or ch != 'u'):
            count_conso += 1
    return count_vowels,count_conso 

print(f"Vowels and Consonants are {count("Pushkar")}")


            
```

```python
#first non repeating characters

from collections import Counter

def firstNonRepeating(s):
    freq = Counter(s)

    for ch in s:
        if freq[ch] == 1:
            return ch
    return None

print(firstNonRepeating("successful"))
```

```python
#check if string is Palindrome

def is_palindrome(s):

    return s == s[::-1]

print(is_palindrome("FaaF"))
```

```python
#check if two strings are sorted
def is_anagram(s1, s2):
    return sorted(s1) == sorted(s2)
print(is_anagram("listen","silent"))
```
