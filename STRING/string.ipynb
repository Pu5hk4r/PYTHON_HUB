```python
#reverse a string 

def reverse(s):
    result = ""

    for ch in s:
        result =  ch + result
    return result

print(reverse("Pushkar"))


```

```python
#find length of string without len function

def string_length(s):
    count = 0
    for _ in s:
        count +=1
    return count
print(f"length of string is {string_length("Pushkar")}")

```

```python
#check if two strings rotation of each other T/C O(N)

def is_rotation(s1, s2):
    if len(s1) != len(s2):
        return False
    return s2 in (s1+s1)

print(is_rotation("abcde", "cdeab"))

# is s2 is rotation of s1
```

```python
#count vowels and consonants in a string
def count(s):
    count_vowels = 0
    count_conso = 0

    for ch in s:
        if(ch == 'a' or ch == 'e' or ch == 'i' or ch =='o' or ch == 'u'):
            count_vowels += 1
        elif(ch != 'a' or ch != 'e' or ch != 'i' or ch !='o' or ch != 'u'):
            count_conso += 1
    return count_vowels,count_conso 

print(f"Vowels and Consonants are {count("Pushkar")}")


            
```

```python
#first non repeating characters

from collections import Counter

def firstNonRepeating(s):
    freq = Counter(s)

    for ch in s:
        if freq[ch] == 1:
            return ch
    return None

print(firstNonRepeating("successful"))
```

```python
#check if string is Palindrome

def is_palindrome(s):

    return s == s[::-1]

print(is_palindrome("FaaF"))
```

```python
#check if two strings are sorted
def is_anagram(s1, s2):
    return sorted(s1) == sorted(s2)
print(is_anagram("listen","silent"))
```

```python
#longest palindromic substring using Brute Force O(N'3)
def long_palindromic_Substr_BF(s):
    longest = ""
    n = len(s)

    for i in range(n):
        for j in range(i, n):
            substring = s[i:j+1]

            if substring == substring[::-1]:
                if len(substring) > len(longest):
                    longest = substring
    return longest


print(long_palindromic_Substr_BF("babad"))
        
    
```

```python
#longest palindrome brute force two sum O(N'3)

def is_palindrome(s, left, right):
    while left < right:
        if s[left] < s[right]:
            return False
        left += 1
        right -= 1
    return True

def longest_palindrome(s):
    n = len(s)
    start = 0
    max_len = 0

    for i in range(n):
        for j in range(i, n):
            if is_palindrome(s,i,j):
                if j - i + 1 > max_len:
                    start = i
                    max_len = j - i + 1
    return s[start : start + max_len]

print(longest_palindrome("babababd"))
```

```python
#longest substring optimal
def expand(s, left, right):
    while left >= 0 and right < len(s) and s[left] == s[right]:
        left -= 1
        right += 1
    return s[left+1:right]

def longest_palindrome(s):
    result = ""
    for i in range(len(s)):
        p1 = expand(s, i, i)      # Odd length
        p2 = expand(s, i, i+1)    # Even length
        result = max(result, p1, p2, key=len)
    return result

print(longest_palindrome("babad"))  # "bab" or "aba"
```

```python
#remove duplicate characters
def remove_duplicate_chars(s):
    seen = set()
    result = ""
    for ch in s:
        if ch not in seen:
            seen.add(ch)
            result += ch
    return result

print(remove_duplicate_chars("programming"))  # "progamin"
```
