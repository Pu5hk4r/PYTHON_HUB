```python
#Birthday cake candle O(N)
def birthdayCakeCandles(candles):
    # Write your code here
    if not candles:
        return -1
        
    mx = float('-inf')
    count = 0
    for i in range(len(candles)):
        if candles[i] > mx:
            mx = candles[i]
            count = 1
        elif candles[i] == mx:
            count += 1
    return count

# second pythonic approach O(N)
def birthdayCakeCandles(candles):
    mx = max(candles)
    return candles.count(mx)      
```
