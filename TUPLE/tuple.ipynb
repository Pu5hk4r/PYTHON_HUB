```python
#accesing tuple elements

t = (1, 2, 3, 4, 5)
print(t[0], t[3], t[-1])
```

```python
# Slice a tuple to get elements from index 2 to 6.
t = (1, 2, 3, 4, 5, 6, 7, 8)
print(t[2:7])  
```

```python
#  Find the length of a tuple.
t = (1, 2, 3, 4, 5)
print(len(t))
```

```python
#Convert a list into a tuple
lst = [1, 2, 3, 4]

t = tuple(lst)
print(t)
```

```python
# Check if 7 exists in a tuple

t = (1,3,5,7,9)
print(7 in t )
```

```python
#Repeat a tuple five times.
t = ("I love father")
print(t * 5)
```

```python
#Count how many times 2 occurs.
t = (2, 4, 2, 6, 2, 8)
print(t.count(2))
```

```python
#Find the index of the first occurrence of 9.
t = (1, 3, 9, 4, 9, 6)
print(t.index(9))
```

```python
#creating a single element tuple

t = (5,)

print(type(t))
```

```python
#concatenate two tuples
t1 = (1, 2, 3)
t2 = (4, 5, 6)

print(t1 + t2)
```

```python
#unpacking a tuple
t = (1, 2, 3)
a,b,c = t
print(a,b,c)
```

```python
#swap first and last elements of a tuple
t = (10, 20, 30, 40, 50)

t = (t[-1],) + t[1:-1] + (t[0],)
print(t)
```

```python
#reverse a tuple
t = (1, 2, 3, 4, 5)
t = t[::-1]
print(t)
```

```python
#convert tuple of characters to a string
t = ('p','y','t','h','o','n')
print(''.join(t))
```

```python
# Nested tuple access.
t = (1, (2, 3), (4, (5, 6)))
print(t[2][1][0]) 
```

```python
# find max and min values
t = (10, 25, 3, 8, 50)

print(max(t), min(t))
```

```python
#sort a tuple
t = (5, 2, 9, 1, 5, 6)
t = tuple(sorted(t))
print(t)
```

```python
# merge two tuples and sort the result
t1 = (1, 2, 3, 4)
t2 = (3, 4, 5, 6)

merged = tuple(set(t1+t2))
print(merged)
```

```python
#convert tuple into dictionary

t = ((1,"one"), (2,"two"), (3,"three"))

d = dict(t)

print(d)
```

```python
#extract the names of student  with grade A

t = (("Alice" , 21, "A"), ("Bob", 22 , "B"), ("charlie", 20 ,"A"))

names = tuple(s[0] for s in t if s[2] == "A")

print(names)
```

```python
#check if all elements are identical

t = (25,25,25,25)

print(all(x == t[0] for x in t))
```

```python
# find common element between tuples
t1 = (1, 2, 3, 4)
t2 = (3, 4, 5, 6)

common = tuple(set(t1) & set(t2))
print(common)
```

```python
#Split list of tuples into two tuples.
lst = [(1,2), (3,4), (5,6)]
t1, t2  = zip(*lst)
print(t1, t2)
```

```python
#find most frequent element
t = (2, 3, 2, 5, 2, 3, 3, 3)

most = max(t, key = t.count)
print(most)
```

```python
#Replace 3rd element (immutability workaround).
t = (10, 20, 30, 40, 50)
lst = list(t)
lst[2] = 99
t = tuple(lst)
print(t)   # (10, 20, 99, 40, 50)
```

```python
#keep only even numbers from a tuple
def even_tuple(t):
    return tuple(x for x in t if x % 2 == 0)

print(even_tuple((1, 2, 3, 4, 5, 6)))   # (2, 4, 6)
```

```python
#Flatten nested tuple.
t = ((1,2),(3,4),(5,6))
flatttend = tuple(x for tup in t for x in tup)
print(flatttend)
```

```python
# tuple for square

t = tuple(x**2 for x in range(1, 11))
print(t)   # (1, 4, 9, ..., 100)
```

```python
#convert  sentence to tuple of words
s = "This is a sample sentence"
t = tuple(s.split())    
print(t)
```
