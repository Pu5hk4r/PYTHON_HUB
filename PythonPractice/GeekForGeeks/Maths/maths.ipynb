```python
#Count digit

def counting(n):
    count = 0
    while n != 0:
        n = n//10
        count += 1
    print(count)
counting(9253)
```

```python
#PAlindrome 

def palindrome(n):
    original = n
    rev = 0
    while n > 0:
        digit = n % 10
        
        rev = rev*10 + digit
        n = n//10
    if original == rev:
        print(f'yes {original} is Palindrome number ')
    else:
        print("Not a Palindrome number")
    
palindrome(12331)
```

```python
#Factorial
def factorial(N):
    flag = 1
    if N==1 or N==0:
        print(1)
    while N > 0 :
        flag = flag * N
        N -= 1
    print(flag)

factorial(5)


    
```

```python
#recursive, O(N) and S(N)
def func(N):
    if N == 1 or N ==0:
        return 1
    return N*func(N-1)
func(20)
```

```python
#GCD of two number--Greatest Common Divisor
#gcd(a,b) <= min(a,b)
# The Greatest Common Divisor (GCD) of two numbers a and b is:

# The largest number that divides both a and b exactly (no remainder).

# Example:
# GCD(12, 18) = 6
# Divisors of 12 → {1, 2, 3, 4, 6, 12}
# Divisors of 18 → {1, 2, 3, 6, 9, 18}
# Common → {1, 2, 3, 6} → max = 6
# 🧠 Brute Force Idea

# Brute force means:

# “Try all possibilities and pick the best one.”

# So for GCD:

# Check all numbers from 1 to min(a, b)
# See which numbers divide both
# Track the largest one


def gcd_bf(a,b):
    gcd =1
    for i in range(1,min(a,b)+1):
        if a % i == 0 and b % i == 0:
            gcd = i
    return gcd

print(gcd_bf(12,18))

# 👉 Time Complexity = O(min(a, b))
# Checking all numbers/divisors, it dont making sense
# A number greater than min(a, b) can never divide both numbers
#to make it little bit efficient we can make reverse loop for i in range(min(a, b), 0, -1):



#
```

```python
#Euclid O(log N) ~ O(log(min(a,b))), S/c O(1)
# gcd(a,b) = gcd(b, a mode b)
# If a number divides both a and b,
# it will also divide a % b.

def euclid(a,b):
    while b!=0:
        a,b = b, a % b
    return a

euclid(12,18)
    
    

```

```python
#LCM BF O(A*B)

def LCM(a,b):
    x = max(a,b)
    while True:
        if x % a == 0 and x % b == 0:
            return x
        x += 1
LCM(12,48)
```

```python
#LCM extension for lcm(a,b,c) O(nlogm)
from math import gcd

def lcm(a,b):
    if a==0 or b==0:
        return 0
    return abs(a*b) // gcd(a,b)

def lcm_array(arr):
    if not arr:
        return 0
    
    result = arr[0]
    for x in arr[1:]:
        result = lcm(result,x)
    return result

#lcm(48,128)
lcm_array([4, 6, 8])
```

```python
# 🧠 What is a Prime Number?
# 👉 A number is prime if:
# It has exactly 2 divisors → 1 and itself
# Prime: 2, 3, 5, 7, 11
# Not prime: 4, 6, 8, 9
#TC O(N)

def prime(n):

    if n <= 1:
        return False
    for i in range(2,n):
        if n % i == 0:
            return False
    return True
prime(23)
```

```python
#optmized PRime
#O(sqrt(N))
import math

def is_prime(n):
    if n<=1:
        return False
    
    for i in range(2,int(math.sqrt(n)) + 1):  #use int otherwise it take float value ,prime no. is nit float
        if n % i == 0:
            return False
    return True

is_prime(29)
```
