```python
#Chunk text for RAG.embeddings,vector DB, LLm pipelines

text = "Python is powerful and easy to learn"

chunk_size = 10

chunk = []

for i in range(0,len(text),chunk_size):
    chunk.append(text[i:i+chunk_size])

print(chunk)
```
